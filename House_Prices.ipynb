```python
import pandas as pd
import numpy as np

df = pd.read_csv('kc_house_data.csv')
df
```

```python
house_df = df[['price', 'bedrooms', 'bathrooms', 'sqft_living', 'sqft_lot', 'floors', 'waterfront', 'view', 'yr_built', 
              'yr_renovated', 'zipcode']]
```

```python
# from sklearn.preprocessing import LabelEncoder
# encoder = LabelEncoder()

# house_df['zipcode']= encoder.fit_transform(df['zipcode'])
# house_df
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
house_df.iloc[:, :] = scaler.fit_transform(house_df.iloc[:, :])
house_df
```

```python
x_columns = 10
X = house_df.iloc[:, 1:x_columns].values
y = house_df.iloc[:, 0].values
y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=3)
```

```python
#all_models = []
all_models = {}
```

```python
from sklearn.linear_model import LinearRegression
model_LR = LinearRegression()
model_LR.fit(X_train, y_train)
y_pred = model_LR.predict(X_test)
score_LR = model_LR.score(X_test, y_test)
print(score_LR)
all_models[model_LR] = score_LR
print(all_models)
```

```python
neighbors = range(1,100)
bestscore_KNN = 0
from sklearn.neighbors import KNeighborsRegressor

for n in neighbors:
    model_KNN = KNeighborsRegressor(n_neighbors = n)
    model_KNN.fit(X_train, y_train)
    score_KNN = model_KNN.score(X_test, y_test)
    #print(score)
    if score_KNN > bestscore_KNN:
        bestscore_KNN = score_KNN
        knn_n = neighbors.index(n)+1
bestscore_KNN

#all_models.append(model_KNN)
all_models[model_KNN] = bestscore_KNN
print(all_models)
```

```python
from sklearn.preprocessing import PolynomialFeatures
scoresPR = []
for i in range(1,6):
    poly=PolynomialFeatures(degree=i)
    x_poly = poly.fit_transform(X)

    
    model = LinearRegression()
    model.fit(X_train, y_train)
    
    score = model.score(X_test, y_test)
    scoresPR.append(score)
    scorePR = max(scoresPR)

    
bestPoly = scoresPR.index(max(scoresPR)) + 1 # location of max score

print('Max polynomial score is: ',scorePR, "with a value of",bestPoly)
```

```python
crit = ['mse', 'mae']
depth = [3, 4, 5, 6]
best_score_DT = 0
from sklearn.tree import DecisionTreeRegressor
for i in crit:
    for d in depth:
        t_model = DecisionTreeRegressor(criterion=i, max_depth = d, splitter='best', random_state=0)
        t_model.fit(X_train, y_train)
        y_pred = t_model.predict(X_test)
        score_DT = t_model.score(X_test, y_test)
    
        if score_DT > best_score_DT:
            best_score_DT = score_DT
print(best_score_DT)
```

```python
estimators = [3, 5, 10, 15]
critera = ['mse', 'mae']
best_score_RF = 0

from sklearn.ensemble import RandomForestRegressor
for i in estimators:
    for c in critera:
        model_RF = RandomForestRegressor(n_estimators = i, criterion = c, random_state = 0)
        model_RF.fit(X_train, y_train)
        y_pred_RF = model_RF.predict(X_test)
        score_RF = model_RF.score(X_test, y_test)
        
        if score_RF > best_score_RF:
            best_score_RF = score_RF
best_score_RF

```
